### Importar csv

```python
### Improtar csv
salarios = pd.read_csv("../Data/linkedin-jobs-usa.csv")
```

### Manejo duplicados

```python
### Eliminar duplicados
salarios.drop_duplicates()
```

```python
salarios.info()
```

```python
###Valores Nulos Metodo no recomendable
df_sin_nulos = salarios.dropna()

df_sin_nulos_columnas = salarios.dropna(axis=1) #Almenos 1 valor nulo
```

```python
##Revision
df_sin_nulos.shape
df_sin_nulos_columnas.shape
```

```python
### Analisis del campo
df_sin_nulos['salary']
```

### Modificación Campo Salario

```python
df_sin_nulos['salary_numerico'] = df_sin_nulos['salary'].replace('[^\d.]', '', regex=True).str.extract('(\d+)').astype(float)
```

```python
df_sin_nulos['salary_numerico'].info()
```

```python
# Crear un boxplot
plt.boxplot(df_sin_nulos['salary_numerico'])

# Añadir etiquetas y título
plt.xlabel('Salarios')
plt.ylabel('Valores en USD')
plt.title('Boxplot')

# Mostrar el gráfico
plt.show()
```

### Se deben sacar todos quienes tengan un salario mayor o igual a 20000

```python
df_sin_nulos = df_sin_nulos[df_sin_nulos['salary_numerico'] >= 20000]
```

### Verificación

```python
# Crear un boxplot
plt.boxplot(df_sin_nulos['salary_numerico'])

# Añadir etiquetas y título
plt.xlabel('Salarios')
plt.ylabel('Valores en USD')
plt.title('Boxplot')

# Mostrar el gráfico
plt.show()
```

### Si se fuera a trabajar con datos de media en vez de eliminar nulos

```python
df_con_relleno = df_sin_nulos.fillna(df_sin_nulos['salary_numerico'].mean())
```

### ELIMINAMOS LA COLUMNA CON NULOS

```python
df_sin_nulos
```

```python
df_sin_nulos.drop('salary', axis=1, inplace=True)
```

```python
df_sin_nulos
```

### Organizar data para exportar

```python
# Agrupar por título y calcular el promedio y la mediana del salario para cada título
df_grouped = df_sin_nulos.groupby('title')['salary_numerico'].agg(['mean', 'median', 'count']).reset_index()

# Renombrar las columnas para mayor claridad
df_grouped.columns = ['title', 'salary_mean', 'salary_median', 'job_count']

# Mostrar el DataFrame resultante
df_grouped
```

### Redondear

```python
df_grouped['salary_mean'] = df_grouped['salary_mean'].round(0).astype(int)
df_grouped['salary_median'] = df_grouped['salary_median'].round(0).astype(int)

df_grouped
```

### Exportar

```python
# Exportar CSV
# Especifica la ruta y el nombre del archivo CSV
ruta_y_nombre_archivo_csv = "export-linkedin-jobs-usa"

# Exporta el DataFrame a un archivo CSV
df_grouped.to_csv("../Data/linkedin-jobs-usa-exported.csv")
```
